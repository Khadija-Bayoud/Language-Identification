```python
languages = ['Arabic', 'English', 'French', 'Russian']
test_df = pd.DataFrame(columns=['file path', 'language'])

for lang in languages:
  path = '/content/drive/MyDrive/Projet/Langues/MFCC/'+lang+'/Preprocessing/'
  df = pd.read_csv(path + lang + '_df.csv')
  test_df = pd.concat([test_df, df], ignore_index=True)

print(test_df.shape)
print(test_df.loc[239])
```
