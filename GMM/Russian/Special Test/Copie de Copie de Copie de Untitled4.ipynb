```python
import numpy as np
from sklearn.mixture import GaussianMixture
import os
from python_speech_features import mfcc
import scipy as sc
```

```python
path = '/content/drive/MyDrive/Projet/Langues/GMM/Russian/Special Test'
files = [filename for filename in os.listdir(path) if filename.endswith('wav')]
files
```

```python
rate, data = sc.io.wavfile.read('/content/drive/MyDrive/Projet/Langues/GMM/Russian/Special Test/RUSS-15secs.wav')
```

```python
mfcc = mfcc(data, rate)
```

```python
energies = np.sum(np.square(mfcc), axis=1)
energies = energies.reshape(-1, 1)

gmm = GaussianMixture(n_components=2, covariance_type='full')
gmm.fit(energies)
means = gmm.means_
silence_component = np.argmin(means)
silence_threshold = means[silence_component] - 2* np.sqrt(gmm.covariances_[silence_component])
non_silent_mask = energies > silence_threshold
mfcc_no_silence = energies[non_silent_mask]

np.savetxt('/content/drive/MyDrive/Projet/Langues/GMM/Russian/Special Test/RUSS-15secs.txt', mfcc_no_silence, delimiter=',')
```
