```python
import os
import joblib
import numpy as np
```

```python
test_files = "/content/drive/MyDrive/Projet/Langues/Decision/Recordings"
```

```python
gmm_ar = joblib.load('/content/drive/MyDrive/Projet/Langues/GMM/Arabic/gmm_arabe64.gmm')
gmm_en = joblib.load('/content/drive/MyDrive/Projet/Langues/GMM/English/gmm_english16.gmm')
gmm_fr = joblib.load('/content/drive/MyDrive/Projet/Langues/GMM/French/gmm_french32.gmm')
gmm_russ = joblib.load('/content/drive/MyDrive/Projet/Langues/GMM/Russian/gmm_russian32.gmm')
#gmm_jp = joblib.load('/content/drive/MyDrive/Projet/Langues/GMM/Japaneese/gmm_japaneese256.gmm')
```

```python
languages = {0 : 'English', 1 : 'Arabic', 2 : 'French', 3 : 'Russian'}
for filename in os.listdir(test_files):
  mfcc = np.loadtxt(os.path.join(test_files, filename), delimiter=',').reshape(-1,1)
  
  score_en = gmm_en.score(mfcc)
  score_ar = gmm_ar.score(mfcc)
  score_fr = gmm_fr.score(mfcc)
  score_russ = gmm_russ.score(mfcc)
  #score_jp = gmm_jp.score(mfcc)

  scores = [score_en, score_ar, score_fr, score_russ]
  print(f"The spoken language in {filename} is {languages[np.argmax(scores)]}")
```
