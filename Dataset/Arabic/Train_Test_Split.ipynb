```python
from pydub import AudioSegment
import os
import shutil
import random
```

```python
dir = "/content/drive/MyDrive/Projet/Langues/Dataset/Arabic"
files_H = os.listdir(dir + ("/H"))
files_F = os.listdir(dir + ("/F"))
print(len(files_H))
print(len(files_F))
```

```python
total_duration_F = sum([AudioSegment.from_file(os.path.join(dir + ("/F"), f)).duration_seconds for f in files_F])
print(total_duration_F/60, 'secs')
```

```python
total_duration_H = sum([AudioSegment.from_file(os.path.join(dir + ("/H"), f)).duration_seconds for f in files_H])
print(total_duration_H/60, 'secs')
```

```python
test_duration = int(total_duration_H * 0.3)
test_repo = "/content/drive/MyDrive/Projet/Langues/Dataset/Arabic/Test/H"
#test_files = []
test_duration_so_far = 0
while test_duration_so_far < test_duration:
    # Randomly select a file from the list of audio files
    filename = random.choice(files_H)
    audio = AudioSegment.from_file(os.path.join(dir + ('/H'), filename))
    file_duration = audio.duration_seconds

    # Add the file to the test set if it won't exceed the test duration
    if test_duration_so_far + file_duration <= test_duration:
        shutil.move(os.path.join(dir + ("/H"), filename), test_repo)
        files_H.remove(filename)
        #test_files.append(filename)
        test_duration_so_far += file_duration
    else :
      break
```

```python
# Durée Test
dir = "/content/drive/MyDrive/Projet/Langues/Dataset/Arabic/Test"
test_H = os.listdir(dir + '/H')
test_F = os.listdir(dir + '/F')

total_duration_H = sum([AudioSegment.from_file(os.path.join(dir + '/H', f)).duration_seconds for f in test_H])
total_duration_F = sum([AudioSegment.from_file(os.path.join(dir + '/F', f)).duration_seconds for f in test_F])

test_duration = (total_duration_H + total_duration_F)/60
print(test_duration)
```
