```python
import os
import pandas as pd
```

```python
df = pd.DataFrame(columns=["file path", "language"])
```

```python
path = "/content/drive/MyDrive/Projet/Langues/MFCC/Russian/Preprocessing/Test"
lang = "Russian"
```

```python
for filename in os.listdir(path + ("/F")) :
    file_path = path + ("/F/") + filename
    df = df.append({"file path": file_path, "language": lang}, ignore_index=True)
```

```python
for filename in os.listdir(path + ("/H")) :
    file_path = path + ("/H/") + filename
    df = df.append({"file path": file_path, "language": lang}, ignore_index=True)
```
